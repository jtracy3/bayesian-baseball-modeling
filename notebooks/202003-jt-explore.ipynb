```python
position_col_names = ['G_p','G_c','G_1b','G_2b','G_3b','G_ss','G_lf','G_cf','G_rf','G_dh']
positions = {
    0: 'P',
    1: 'C',
    2: '1B',
    3: '2B',
    4: '3B',
    5: 'SS',
    6: 'LF',
    7: 'CF',
    8: 'RF',
    9: 'DH'
}
player_batting_appearances.loc[:, position_col_names]
```

```python
pos_max = np.argmax(player_batting_appearances.loc[:, position_col_names].values, axis=1)
pos_cnts = np.bincount(pos_max)
for k, v in positions.items():
    print(f'{v} : {pos_cnts[k]} player years')
```

```python
def get_main_position(row):
    pos_idx = np.argmax(row)
    return positions[pos_idx]
```

```python
player_batting_appearances['pos_main'] = player_batting_appearances.apply(lambda row: get_main_position(row[position_col_names]), axis=1)
```

```python
player_batting_appearances_wo_p = player_batting_appearances.loc[player_batting_appearances.pos_main != 'P'].copy()
```

```python
player_batting_appearances_wo_p = player_batting_appearances_wo_p.sort_values(['playerID', 'yearID'])

park_clean = []
for index, player_year in player_batting_appearances_wo_p.groupby(['playerID', 'yearID']):
    N = len(player_year)
    
    if N == 1:
        park_clean.append(player_year.park.values[0])
        continue
        
    max_G = player_year.G.max()
    max_AB = player_year.AB.max()
    
    if len(player_year[player_year.G == max_G]['park']) > 1:
        for i in range(N):
            park_clean.append(player_year[player_year.AB == max_AB]['park'].values[0])
    else:
        for i in range(N):
            park_clean.append(player_year[player_year.G == max_G]['park'].values[0])
    
player_batting_appearances_wo_p['park_clean'] = park_clean
```
